```python
df_4 = pd.read_csv('step_4_lesson_7.csv')
```

```python
types_15th = df_4.groupby(['date', 'browser', 'user_type'], as_index = False) \
.agg({'visits': 'sum'}) \
.sort_values('visits', ascending = False) \
.reset_index(drop = True)
```

```python
types_15th.head()
```

```python
# объединяю свою таблицу с таблицей другого аналитика
merge_df = df.merge(types_15th, on = ['date', 'browser', 'user_type'])
```

```python
# считаю количество визитов на человека (отношение visits к users) в разрезе по браузеру и значениям Robots/People
merge_df['us_vis'] = round(merge_df['visits'] / merge_df['users'], 0)
```
