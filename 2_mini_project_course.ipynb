```python
# импортирую pandas и считываю файл csv
import pandas as pd 
pd.set_option('display.max_columns', None)    #применяю метод, чтобы все столбцы таблицы были показаны
bookings = pd.read_csv('2_bookings.csv', encoding = 'windows = 1251', sep = ';')
```

```python
# знакомлюсь с размером таблицы
bookings.shape
```

```python
# смотрю, какие типы данных присутствуют в столбцах таблицы
bookings.dtypes
```

```python
# привожу названия столбцов к единому стилю
bookings = bookings.rename(columns = {'Hotel': 'hotel',
                          'Is Canceled': 'is_canceled',
                          'Lead Time': 'lead_time',
                          'arrival full date': 'arrival_full_date',
                           'Arrival Date Year': 'arrival_date_year',
                           'Arrival Date Month': 'arrival_date_month',
                           'Arrival Date Week Number': 'arrival_date_week_number',
                           'Arrival Date Day of Month': 'arrival_date_day_of_month',
                           'Stays in Weekend nights': 'stays_in_weekend_nights',
                           'Stays in week nights': 'stays_in_week_nights',
                           'stays total nights': 'stays_total_nights',
                            'Adults': 'adults',
                            'Children': 'children',
                            'Babies': 'babies',
                            'Meal': 'meal',
                            'Country': 'country',
                            'Reserved Room Type': 'reserved_room_type',
                            'Assigned room type': 'assigned_room_type',
                            'customer type': 'customer_type',
                            'Reservation Status': 'reservation_status',
                            'Reservation status_date': 'reservation_status_date'})
```

```python
bookings
```

```python
# анализирую, пользователи из каких 5 стран совершили наибольшее число успешных бронирований
bookings[bookings['is_canceled'] == 0] \
.groupby('country', as_index = False) \
.agg({'is_canceled': 'count'}) \
.sort_values('is_canceled', ascending = False) \
.head()
```

```python
# анализирую, на сколько ночей в среднем бронируют отели типа City Hotel и Resort Hotel
bookings.groupby(['hotel'], as_index = False) \
.agg({'stays_total_nights': 'mean'}).round(2) \
.sort_values('stays_total_nights', ascending = False)
```

```python
# анализирую, сколько раз тип номера, присвоенного клиенту (assigned_room_type), отличается от изначально забронированного (reserved_room_type)
bookings.query('assigned_room_type != reserved_room_type').shape[0]
```

```python
# анализирую, на какой месяц чаще всего оформляли бронь в 2016м году
bookings.query('arrival_date_year == 2016') \
.groupby('arrival_date_month', as_index = False) \
.agg({'is_canceled': 'count'}) \
.sort_values('is_canceled', ascending = False)
```

```python
# анализирую, изменился ли самый популярный месяц в 2017м году
bookings.query('arrival_date_year == 2017') \
.groupby('arrival_date_month', as_index = False) \
.agg({'is_canceled': 'count'}) \
.sort_values('is_canceled', ascending = False)
```

```python
# анализирую, на какой месяц чаще всего оформляли бронь с 2015 по 2017 гг
bookings.groupby(['arrival_date_year', 'arrival_date_month'] , as_index = False) \
.agg({'is_canceled': 'count'}) \
.sort_values(['is_canceled', 'arrival_date_year'], ascending = False)
```

```python
# более караткий вариант для нахождения самого ппопулярного месяца
bookings.groupby('arrival_date_year', as_index = False) \
.agg({'arrival_date_month': pd.Series.mode}) 
```

```python
# анализирую, на какой месяц (arrival_date_month) бронирования отеля типа City Hotel отменялись чаще всего в 2015, 2016, 2017
bookings.query('(is_canceled == 1) & (hotel == "City Hotel")') \
.groupby('arrival_date_year', as_index = False) \
.agg({'arrival_date_month': pd.Series.mode}) \
.sort_values('arrival_date_month', ascending = False)
```

```python
# анализирую, какая колонка имеет наибольшее среднее значение
bookings[['adults', 'children', 'babies']].mean().idxmax()
```

```python
# Создаю колонку total_kids, объединив столбцы children и babies, и 
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
# ванализирую, для отелей какого типа среднее значение переменной оказалось наибольшим
# 1 способ
bookings.groupby('hotel', as_index=False)['total_kids'].mean().round(2).max()
```

```python
#2 способ
bookings.groupby('hotel', as_index=False) \
.agg({'total_kids': 'mean'}).round(2).max()
```

Churn rate (отток, коэффициент оттока) – это процент подписчиков (например, на push-уведомления от сайта), которые отписались от канала коммуникации, отказались от услуг сервиса в течение определенного периода времени. Иными словами, представляет собой отношение количества ушедших пользователей к общему количеству пользователей, выраженное в процентах.
В нашем случае Churn Rate - это процент клиентов, которые отменили бронирование. Давайте посмотрим, как эта метрика связана с наличием детей у клиентов!

Создайте колонку has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка (total_kids), в противном случае – False. Далее проверьте, среди какой группы пользователей показатель оттока выше. 

В качестве ответа укажите наибольший % оттока, округленный до 2 знаков после точки (то есть доля 0.24563 будет 24.56% и в ответ пойдёт 24.56)

```python
# создаю столбец "has_kids", состоящую из булевых значений, указывающих, есть ли дети (True) или нет (False) для каждой строки
bookings['has_kids'] = bookings['total_kids'] > 0
```

```python
# считаю, какой коэффициент churn rate у клиентов с детьми:
churn_has_kids = round((bookings.query('is_canceled == 1 & has_kids == True').shape[0] / bookings.query('has_kids == True').shape[0]) * 100, 2)
```

```python
churn_has_kids
```

```python
# считаю, какой коэффициент churn rate у клиентов без детей:
churn_has_nokids = round((bookings.query('is_canceled == 1 & has_kids == False').shape[0] / bookings.query('has_kids == False').shape[0]) * 100, 2)
```
